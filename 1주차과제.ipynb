- _PAPAGO_(__번역__)

몇 년 전 일본 여행을 갈려고 일본에서 숙소를 구하려고 했다. 일본인 호스트한테 메일을 보내야 하기도 했고 그 사람한테 안내를 받아야 해서 _파파고_를 사용해 번역을 했다.일본어에서 한국어로 번역한 결과물은 사람이 쓴 것 같다는 느낌이 잘 안 들어 어색했다. 한국어와 문법이 비슷한 일본어도 번역을 못하는 것을 느끼고 그 뒤부터 잘 사용하지 않았다. 몇 년 뒤 올해 학교 영어 과제를 해야 할 때가 있어 오랜만에 _파파고_를 사용했다. 영어를 한국어로 번역한 결과물을 보니 자연스럽게 읽히고 동음이의어도 구분했다. 또한 상품이름이나 회사 이름 같은 단어도 그 단어 자체를 해석하지 않고 단어를 그대로 출력했다. 많이 발달했다는 생각이 들었다. 찾아보니 원인은 두 개가 있었다. 첫째로 알고리즘이 발달되었다. 처음에는 RBMT(규칙기반기계번역)를 사용했다. 두 언어의 문법 규칙을 기반해서 개발되었다. 문법 규칙 추출이 어려웠고 번역 언어 확장도 어려웠다. 그다음에는 SMT(통계 기반 기계번역)를 사용했다.입력 문장의 단어, 구, 절을 보고 번역을 했고 번역사전을 사용해 확률에 기반했다. 어순이 같은 언어 끼리는 성능이 괜찮았지만 어순이 다르면 성능이 떨어졌다. 드디어 NMT(인공신경망 기계 번역)이 등장했다. 단어를 고차원 벡터로 처리했다. 단어의 입력, 출력 순서를 학습했고 문장 전체의 의미를 파악한 후 번역을 수행했으며 그로 인해 동음이의어도 전체 문장의 의미에 맞게 번역이 가능해졌다. 전보다 매끄러운 번역 문장이 생성됐다. _파파고_는 처음부터 인공신경망을 이용한 NSMT(NAVER Statistical Machine Translation) 기반으로 번역했고 그 이후 현재도 사용하는 N2MT를 이용해 번역한다. 동음이의어 구분과 번역 정확도가 나아졌다. 두 번째는 사용자 평가를 이용한 피드백이다. 피드백을 통해 딥러닝으로 번역이 점점 좋아지고 있다.

- _VITO_(__음성__)

지인분한테 컴퓨터 쪽 관련해서 일하고 싶다고 하니까 알려준 음성인식 어플이다. '리턴제로'스타트업에서 개발했다. _VITO_는 우리가 통화 녹음을 하면 자동으로 녹음된 내용을 텍스트로 변환한다. 또 자신과 상대방의 목소리를 구분하는 화자 분리 기술도 적용돼있다. _아마존의 알렉사_ 같은 기존의 음성인식기술에서는 집에서의 생활에서 특정 패턴의 발화 분석에 최적화되어있다. _VITO_는 집, 직장, 학교 등 일상생활에서 많이 사용되는 기능을 가진 어플이어서 실제 생활에 쓰이는 다양한 발화를 분석한다. 처음에는 내 목소리를 잘 분석하지 못할 수 있다. 하지만 계속 시도할수록 나의 말투와 언어습관 방언 그리고 자주 쓰는 어휘나 문구를 분석해 머신러닝을 한다. 이로 인해 점점 더 정확도가 향상된다. _VITO_를 이용함으로써 중요한 회의 내용을 정확한 텍스트로 한눈에 살펴볼 수 있으며 법정 증거가 될 수 있다.

- _꽃 이름 찾기_(__이미지__)

요즘 보는 드라마에서 등장인물이 꽃을 사진 찍어 이름을 알 수 있는 장면을 봤다. _다음_ 앱에서 꽃을 사진 찍으면 꽃의 이름을 말해주는 서비스이다. 국내에서 주로 파는 400여 가지 꽃 품종을 분석할 수 있다. 이 품종들의 10만 장 정도를 데이터로 활용했다. 이 10만장의 데이터는 아주 많은 데이터에서 데이터 마이닝을 통해 적절한 것을 추출한 데이터다. 이미지에 있는 각각의 픽셀을 분석해 학습하고 새로운 이미지의 각각의 픽셀 데이터와 비교해 꽃 종류를 분류한다. 학습과정은  딥러닝을 이용하고 정확히 품종을 구분할 수 있다.

- _DRIVE WISE_(__자율주행__)

_기아_에서 사용하는 ADAS(Advanced Driver Assistance System) 일명 첨단 운전자 보조 시스템 기술의 브랜드이다. 자율 주행의 기술은 adas 기술과 커넥티드 기술의 통신 기반의 기술의 집합체이다. _DRIVE WISE_는 자동 긴급제동 시스템인 AEB ( Autonomous Emergency Braking) 기술을 이용해 위험 긴급상황일 때 운전자가 제동장치를 밟지 않아도 스스로 속도를 줄이거나 멈추어 안전한 운전이 되게 한다. 그 외에도 주행 조향 보조 시스템인 LKAS (Lane Keep Assist System)는 차선을 벗어날 때 방향을 자동으로 조절해서 차선을 유지할 수 있다. ASCC (Advanced Smart Cruise Control) 기술을 이용해 운전자가 도로교통법에 위반되지 않도록 정해 놓은 속도로 달리고 앞차와 간격을 자동으로 유지한다. 후측방 충돌 회피 지원 시스템인 ABSD( Active Blind Spot Detection) 사각지대 충돌 위험을 감지해 안전한 차로 변경을 돕는다. AVM(Around View Monitor) 기술을 이용해 차량 주변 상황을 시각적으로 보여준다. 이렇게 꼼꼼하게 구성된 ADAS 기술을 이용해 최고의 안정성을 보장한다. ADAS 기술은 모든 주변 상황을 거대한 데이터로 받아 분석을 함으로써 인간의 오감보다 빠른 속도와 정확성으로 반응할 수 있으며 미치지 않는 영역까지 분석해 사고의 위험을 현저히 낮춘다. 편의성 영역에서는 인간의 시각이 닿지 않는 차량의 후방 또는 사각지대를 볼 수 있어 주차를 더 깔끔하게 할 수 있게 되었다. 이처럼 자율 주행기술은 음성인식과 이미지 인식 기술이 바탕이 되며 따라서 같이 발전하고 있다. 나는 자율 주행기술이 인공지능기술의 집합체라는 생각이 든다.

- 마무리

인공지능이 발전함에 따라 무조건 같이 나아져야 하는 분야가 있는데 바로 인공지능 관련 법률이다. 인공지능이 판단하에 자율 주행을 하다가 유인 자동차와 사고가 나게 되면 사고의 책임자가 누구냐는 의문이 따른다. 인공지능은 완벽하니 100% 유인 자동차의 운전자 잘못일까? 인공지능에도 분명 허점이 있을 것이므로 자율 주행 자동차를 만든 회사의 잘못인가? 자율 주행 자동차의 잘못인가? 아직 이런 분쟁을 해결할 법률이 우리나라에는 없다.
거의 모든 자동차 회사에서 몇 년 내로 자율 주행 자동차를 상용화한다고 발표했는데도 말이다. 자율 주행과 관련된 법률 제정이 시급하며 어쩌면 인공지능 기술의 발전보다 더 중요한 문제일 수도 있다. 나는 이 질문과 자율 주행은 인공지능의 집합체라고 생각해 자율 주행과 관련된 인공지능 기술에 흥미를 느끼게 되었다. 광 주인공 지능 사관학교를 통해 인공지능 전문 인력이 돼서 하루빨리 자율 주행 자동차를 상용화 시키고 싶다. 또한 인공지능 관련 지식을 빨리 습득해 법률을 제정할 때 많은 도움을 주고 싶다.
